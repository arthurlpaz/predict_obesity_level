### Construindo o modelo de aprendizado

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
```

```python
len(df_obesity)
```

```python
df_obesity.columns
```

```python
X_obesidade = df_obesity[ ['Gender', 'Age', 'Height', 'Weight', 'family_history_with_overweight',
       'FAVC', 'FCVC', 'NCP', 'CAEC', 'SMOKE', 'CH2O', 'SCC', 'FAF', 'CALC',
       'MTRANS'] ]
```

```python
X_obesidade.head()
```

```python
Y_obesidade = df_obesity['NObeyesdad']
Y_obesidade.head()
```

### Divisão base de treinamento e teste

```python
from sklearn.model_selection import train_test_split

X_treino, X_teste, Y_treino, Y_teste = train_test_split(X_obesidade, Y_obesidade, test_size=0.20, random_state=0)
```

```python
X_treino.shape, X_teste.shape
```

```python
Y_treino.shape, Y_teste.shape
```

```python
from sklearn.impute import SimpleImputer

# Criar um imputer para preencher valores ausentes com a média
imputer = SimpleImputer(strategy='mean')

# Aplicar a imputação nos rótulos (Y_treino)
Y_treino = imputer.fit_transform(Y_treino.values.reshape(-1, 1))
```

```python
modelo.fit(X_treino, Y_treino)
```
